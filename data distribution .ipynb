```python
import os 
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
import imagesize

width, height = imagesize.get("test.png")
print(width, height)
```

```python
path = '/home/minjun/Jupyter/data/ocr/2017_MLT/test/'
columns=['width','height']
df = pd.DataFrame(columns=columns)
print(df)
```

```python
for p, d, f in os.walk(path):
    for i in range(len(f)):
        if f[i].endswith('jpg') or f[i].endswith('JPG') :
            img = cv2.imread(p+f[i])
            each_df = pd.Series([img.shape[0],img.shape[1]],index=columns)
            df = df.append(each_df,ignore_index=True)
        if i % 100==0:
            print(i)
            
            
        
```

```python
df['max'] = df[columns].max(axis=1)
```

```python
df['max'].hist(bins=100)
```
